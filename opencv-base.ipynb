```python
!pip3 install opencv-python

import cv2
```

```python
img = cv2.imread('sample.jpg')
print(type(img))

if type(img) == None:
    raise Exception("Image path is incorrect!")
```

```python
plt.imshow(img)
```

```python
fixed_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_image)
```

```python
gray_image = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_image, cmap='gray')
```

```python
# re-size the image

resized_image = cv2.resize(fixed_image, (200,200))
plt.imshow(resized_image)
print(resized_image.shape)
```

```python
# re-size image by percentage

w_ratio = 0.5
h_ratio = 0.5

new_image = cv2.resize(fixed_image, (0,0), fixed_image, w_ratio, h_ratio)
plt.imshow(new_image)
```

```python
# flip image
fliped_image = cv2.flip(fixed_image, 0)
plt.imshow(fliped_image)
```

```python
# write image to a file
cv2.imwrite('new_gray_image.jpg', gray_image)
```
